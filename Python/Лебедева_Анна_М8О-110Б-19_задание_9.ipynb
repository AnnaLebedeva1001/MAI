```python
#Задание №1
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 6)
data_1 = np.random.randint(2, 15, size = 5)
data_2 = np.random.randint(3, 20, size = 5)

fig,ax = plt.subplots()

ax.bar(x, data_1, color='silver')
ax.bar(x, data_2, bottom = data_1, color='pink')
plt.ylabel('Количество оценок среди парней и девушек')
ax.legend(labels = ('парни', 'девушки'), loc = 'upper center')

plt.ylim(0, 35)

plt.show()
```

```python
#Задание №2
plt.figure(figsize=(8, 6), dpi=80)
plt.axes([0,0,1,1])

x = [33, 33, 33]
y = [11, 22, 13, 20, 5, 28]
offset=0.4
colors1=['salmon','pink','beige']
colors2=['silver', 'violet']
plt.pie(x, radius=1, colors=colors1, labels=('Арбузы', 'Груши', 'Бананы'), wedgeprops=dict(width=offset, edgecolor='w'))
plt.pie(y, radius=1-offset, colors=colors2, wedgeprops=dict(width=offset, edgecolor='w'))
plt.show()
```
