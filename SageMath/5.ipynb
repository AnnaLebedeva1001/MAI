<h1 style='color: black'>Задание №5. Алгебраические уравнения 4-й степени</h1>

```python
var("x")
poly_x = -5*x**4 + 2*x**3 - 3*x**2 +7*x - 11
show(poly_x)
```

```python
poly_x /= poly_x.coefficient(x, 4)
show(poly_x)
```

```python
# замена x = y - a/4, где a - коэф. при x**3
a = poly_x.coefficient(x, 3)
show("a = ", a)
```

```python
var("y")
poly_y = poly_x(x = y - (a/4)).expand().simplify_full()
show(poly_y)
```

```python
poly_y /= poly_y.coefficient(y, 4)
show(poly_y)
```

```python
pqr = {'p': poly_y.coefficient(y, 2)
       , 'q': poly_y.coefficient(y, 1)
       , 'r': poly_y.coefficient(y, 0)}
show(pqr)
```

```python
var("s p q r")
poly_s = 2*s**3 - p*s**2 - 2*r*s + r*p - q**2/4
show(poly_s)
```

```python
poly_s_n = poly_s(**pqr)
show(poly_s_n)
```

```python
sols = solve(poly_s_n, s)
```

```python
for sol in sols:
    show(sol.rhs().n(digits=3))
```

```python
# Выбираем любой s, не равный нулю
s_0 = sols[2].rhs()
show(s_0.n(digits=3))
```

```python
var("y s p q")
poly_y_1 = y**2 - y*sqrt(2*s - p) + q/(2*sqrt(2*s - p)) + s
poly_y_2 = y**2 + y*sqrt(2*s - p) - q/(2*sqrt(2*s - p)) + s
```

```python
show(poly_y_1)
show(poly_y_2)
```

```python
show(s_0)
```

```python
poly_y_1_n = poly_y_1(**pqr, s=s_0)
poly_y_2_n = poly_y_2(**pqr, s=s_0)
show(poly_y_1_n.simplify_full())
show(poly_y_2_n)
```

```python
sols = solve(poly_y_1_n, y)
```
